```python
import math
import matplotlib.pyplot as plt

def plot_roc_grid(
    results,
    models_to_plot=("logistic_regression", "random_forrest", "svm"),
    group_keys=("antibiotic", "X_encoding"),
    ncols=3,
    figsize_per_ax=(5.2, 4.6),
):
    """
    results: list[dict] where each dict has at least:
      - model, antibiotic, X_encoding
      - fpr (array-like), tpr (array-like)
      - roc_auc_mean, roc_auc_std (optional but recommended)
    """

    # ---- group results by (antibiotic, X_encoding) ----
    groups = {}
    for r in results:
        if r.get("model") not in models_to_plot:
            continue
        if "fpr" not in r or "tpr" not in r:
            continue

        key = tuple(r.get(k) for k in group_keys)
        groups.setdefault(key, []).append(r)

    # Sort panels in a nice order: antibiotic then encoding (you can change this)
    def sort_key(k):
        ab, enc = k
        # put common encodings in a stable order if you want:
        enc_rank = {"label": 0, "Label": 0, "OneHot": 1, "onehot": 1, "FCGR": 2, "fcgr": 2}
        return (str(ab), enc_rank.get(str(enc), 999), str(enc))

    panel_keys = sorted(groups.keys(), key=sort_key)

    n_panels = len(panel_keys)
    if n_panels == 0:
        raise ValueError("No panels to plot. Check models_to_plot / keys / presence of fpr,tpr.")

    nrows = math.ceil(n_panels / ncols)

    figsize = (figsize_per_ax[0] * ncols, figsize_per_ax[1] * nrows)
    fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=figsize, sharex=True, sharey=True)
    axes = axes.ravel() if hasattr(axes, "ravel") else [axes]

    for i, key in enumerate(panel_keys):
        ax = axes[i]
        ab, enc = key
        items = groups[key]

        # Keep model order consistent within each panel
        items_sorted = sorted(items, key=lambda d: models_to_plot.index(d["model"]))

        for r in items_sorted:
            auc_mean = r.get("roc_auc_mean", None)
            auc_std  = r.get("roc_auc_std", None)

            if auc_mean is not None and auc_std is not None:
                label = f'{r["model"]} (AUC={auc_mean:.3f}±{auc_std:.3f})'
            elif auc_mean is not None:
                label = f'{r["model"]} (AUC={auc_mean:.3f})'
            else:
                label = f'{r["model"]}'

            ax.plot(r["fpr"], r["tpr"], label=label)

        ax.plot([0, 1], [0, 1], "k--", linewidth=1)
        ax.set_title(f"{ab} + {enc}")
        ax.set_xlim(0, 1)
        ax.set_ylim(0, 1)
        ax.set_xlabel("False Positive Rate")
        ax.set_ylabel("True Positive Rate")
        ax.legend(fontsize=8, loc="lower right")

    # Turn off unused axes if you have < nrows*ncols panels
    for j in range(n_panels, len(axes)):
        axes[j].axis("off")

    plt.tight_layout()
    plt.show()


plot_roc_grid(
    cv_results_final,
    models_to_plot=("logistic_regression", "random_forrest", "SVM"),
    group_keys=("antibiotic", "X_encoding"),
    ncols=3,
)
```
